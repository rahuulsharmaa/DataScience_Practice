Calculating Confidence Interval for avg fair in Titanic dataset using t-procedure(assuming that we dont know the population standard deviation)

```python
import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('Datasets/titanic.csv')
```

```python
# Split into train and test (80% train, 20% test)
train_df, test_df = train_test_split(df, test_size=0.2, random_state=42)

# Show shapes
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
```

```python
df=pd.concat([train_df.drop(columns=['Survived']), test_df]).sample(418)
```

```python
df['Fare'].plot(kind='kde')
```

```python
# sample size=30--->10 times

samples=[]
stds=[]
for i in range(10):
    x=df['Fare'].dropna().sample(30).values
    stds.append(x.std())
    samples.append(x.tolist())
```

```python
samples=np.array(samples)
```

```python
sampling_means=np.mean(samples,axis=1)
```

```python
sampling_std=np.mean(stds)
```

```python
import seaborn as sns
```

```python
sns.kdeplot(sampling_means,fill=True)
```

```python
# For 95% confidence interval
lower_limit=sampling_means.mean()-2.042*sampling_std/np.sqrt(30)
upper_limit=sampling_means.mean()+2.042*sampling_std/np.sqrt(30)
```

```python
print('The range is from',lower_limit,'to',upper_limit)
```
