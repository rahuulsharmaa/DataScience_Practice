BINOMIAL DISTRIBUTION

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
n=10 #nuumber of trials
p=0.8 #probability of success
size=1000 #number of samples to generate[activity of doing coin toss 10 times is repeated 1000 times]

binomial_dist=np.random.binomial(n, p, size)
plt.hist(binomial_dist, density=True)
plt.show()

binomial_dist.shape#[if you see 6 it means in 6 out of 10 tosses we got head]
#if probability of success is low
```

  central limit theorem

```python
#setting parameters
num_samples = 1000
sample_size = 30
distribution_range = (0, 1)

#Generate samples from a uniform distribution
sample=np.random.uniform(distribution_range[0], distribution_range[1], (num_samples, sample_size))

#  1000 samples each of size 30

#Calculate the mean of each sample
sample_means=np.mean(sample, axis=1)

plt.hist(sample_means, bins=30, density=True,edgecolor='black')
plt.title('Distribution of Sample Means from Uniform Distribution')
plt.xlabel('Sample Mean')   
plt.ylabel('Density')
plt.show()    
```

Checking with Exponential Distribution

```python
num_samples = 10000
sample_size = 500
lambda_param = 2

#Generate samples from an exponential distribution
samples=np.random.exponential(scale=1/lambda_param, size=(num_samples, sample_size))

#Calculate the mean of each sample
sample_means=np.mean(samples, axis=1)

plt.hist(sample_means, bins=30, density=True, edgecolor='black')
plt.title('Distribution of Sample Means from Exponential Distribution')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

CHECKING OTHER DISTRIBUTIONS

```python
#Setting parameters
num_samples = 1000
sample_size = 50

#Poisson distribution parameters
poison_lambda = 5

#Gama distribution parameters
gamma_shape = 2
gamma_scale = 1

#Binomial distribution parameters
binomial_n = 10
binomial_p = 0.5

#Generate samples from Distributions
poisson_samples = np.random.poisson(lam=poison_lambda, size=(num_samples, sample_size))
gamma_samples = np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))
binomial_samples = np.random.binomial(n=binomial_n, p=binomial_p, size=(num_samples, sample_size))

#Calculate the mean of each sample
poisson_means = np.mean(poisson_samples, axis=1)
gamma_means = np.mean(gamma_samples, axis=1)
binomial_means = np.mean(binomial_samples, axis=1)

#Plotting the histograms of sample means
fig, axs = plt.subplots(3, 1, figsize=(8, 12))

#Poisson Distribution
axs[0].hist(poisson_means, bins=30, density=True, edgecolor='black')
axs[0].set_title('Distribution of Sample Means from Poisson Distribution')
axs[0].set_xlabel('Sample Mean')
axs[0].set_ylabel('Density')

#Gamma Distribution
axs[1].hist(gamma_means, bins=30, density=True, edgecolor='black')
axs[1].set_title('Distribution of Sample Means from Gamma Distribution')
axs[1].set_xlabel('Sample Mean')
axs[1].set_ylabel('Density')

#Binomial Distribution
axs[2].hist(binomial_means, bins=30, density=True, edgecolor='black')
axs[2].set_title('Distribution of Sample Means from Binomial Distribution')
axs[2].set_xlabel('Sample Mean')
axs[2].set_ylabel('Density')
plt.tight_layout()
plt.show()
```

CHECKING IT BY GAMA DISTRIBUTION

```python
import numpy as np

#set the parameters
num_samples = 10000
sample_size = 50

#Gama distribution parameters
gamma_shape = 2
gamma_scale = 1

#Calculate theoretical mean and variance
theoretical_mean = gamma_shape * gamma_scale
theoretical_variance = gamma_shape * (gamma_scale ** 2)

#Generate samples from a gamma distribution
samples = np.random.gamma(shape=gamma_shape, scale=gamma_scale, size=(num_samples, sample_size))

#Calculate the mean of each sample
sample_means = np.mean(samples, axis=1)

#Calculate empirical mean and variance of sample means
empirical_mean = np.mean(sample_means)
empirical_variance = np.var(sample_means)

#Compare theoretical and empirical results
print(f"Theoritical Mean: {theoretical_mean:.4f}")
print(f"Empirical Mean: {empirical_mean:.4f}")
print(f"\n")
print(f"Theoritical Variance: {theoretical_variance/sample_size:.4f}")
print(f"Empirical Variance: {empirical_variance:.4f}")
```
