Using QQ plot to check the normality of distribution

```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df=pd.read_csv('Datasets/iris.csv')
```

```python
df.head()
```

```python
sns.kdeplot(df['Sepal.Length'])
```

```python
temp=df['Sepal.Length'].sort_values().values
temp
```

```python
y_quant=[]
for i in range(1,101):
    y_quant.append(np.percentile(temp,i))
```

```python
samples=np.random.normal(loc=0,scale=1,size=1000)
```

```python
x_quant=[]
for i in range(1,101):
    x_quant.append(np.percentile(samples,i))
```

```python
sns.scatterplot(x=x_quant,y=y_quant)
```

Using QQ plot to check if distribution is uniform or not

```python
import scipy.stats as stats
```

```python
sample_data=np.random.uniform(low=0,high=1,size=1000)
```

```python
plt.hist(sample_data)
```

```python
#Fit a uniform distribution to the data:
# This finds the best uniform distribution that matches your data.
# A uniform distribution needs 2 things:

# loc  starting point

# scale  width (range)

params=stats.uniform.fit(sample_data)

theoritical_dist=stats.uniform(loc=params[0],scale=params[1])

```

```python
#Create a QQ plot of the data using the uniform distribution
#Default is normal distribution
fig=sm.qqplot(sample_data,dist=theoritical_dist,line='45')
```
