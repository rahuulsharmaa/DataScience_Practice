Pareto Distribution

```python
import numpy as np
import matplotlib.pyplot as plt
alpha=1
x_min=1
x=np.linspace(0,5,100)

y=alpha*x_min**alpha/x**(alpha+1)

plt.plot(x,y)
```

```python
#Create log-log plot
plt.plot(np.log(x),np.log(y))

plt.xlabel('log(x)')
plt.ylabel('log(y)')
plt.title('Log-Log Plot of Power-Law Distribution') 
```

BY QQ plot

```python
import scipy.stats as stats
import statsmodels.api as sm
alpha=2
xm=1

#Generate set of random data from Pareto distribution
sample_data=stats.pareto.rvs(b=alpha,scale=xm,size=1000)
```

```python
plt.hist(sample_data)
```

```python
#Fits pareto distribution to the data
params=stats.pareto.fit(sample_data,floc=0)

#Create a Pareto distribution object with the fitted parameters
theoritical_dist=stats.pareto(b=params[0],scale=params[2])
```

```python
#Create a QQ plot of the data using pareto distribution

fig=sm.qqplot(sample_data,dist=theoritical_dist,line='45')
```
